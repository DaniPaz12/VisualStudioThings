### Calcula el módulo de un vector de $n$ elementos, introducidos por teclado.

```python
import cmath
lista=[]
suma = 0
numero=int(input("Introduce los elementos del vector: "))
while numero !=0:
  lista.append(numero)
  numero=int(input("Introduce los elementos del vector: "))
  suma += numero**2
print(f"El módulo de los elementos es ...{cmath.sqrt(suma)}")


```

### Elabora un código que rellene de forma aleatoria una lista de 10 elementos y luego la ordene.
Existen multitud de fuentes en Internet con la solución en Python del algoritmo Bubble Sort, resiste la tentación y hazlo por ti mismo.



```python
import random
lista_numeros=[]#creamos la lista
for i in range(10):#que sea en el rango de 9 números
  nu=random.randrange(1,20)#que escoja aleatoriamente entre los números 1 al 20 (le podemos dar otros valores si queremos)
  lista_numeros.append(nu)#agregamos esos números a la lista_numeros

print(lista_numeros)#nos enseña la lista de números creada
lista_ordenada=[]#creamos una lista donde estarán los números ordenados
while lista_numeros:
  mini=min(lista_numeros)#creamos la variable mini donde la variable 'min' nos devolverá el elemento más pequeño
  lista_ordenada.append(mini)#agregamos los valores de mini en la lista ordenada
  lista_numeros.remove(mini)#eliminamos los valores que nos va dando la variable mini de nuestra primera lista (lista_ordenada)
print(lista_ordenada)#una vez haya recorrido todos los elementos de la lista_ordenada, nos mostrará la nueva lista
```
